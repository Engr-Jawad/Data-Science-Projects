# Here's a practical project-style problem that combines NumPy, Pandas, and Matplotlib.

Problem: Monthly Sales Analysis

You are given the monthly sales data (in dollars) of three products.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
product_a = [1200, 1350, 1400, 1250, 1500, 1600, 1700, 1650, 1800, 1900, 2000, 2100]
product_b = [1000, 1100, 1150, 1200, 1250, 1300, 1400, 1450, 1500, 1600, 1700, 1800]
product_c = [800, 850, 900, 950, 1000, 1100, 1150, 1200, 1250, 1300, 1400, 1500]
# Convert all sales lists into NumPy arrays.
convert_to_nparr = np.array([product_a,product_b,product_c])
print("lists   are convert to numpy array : \n", convert_to_nparr)
print("\n\n")
# Calculate the total yearly sales for each product.
yearly_sale = np.sum([product_a,product_b,product_c],axis=1)
print("yearly_sale of the three products : ", yearly_sale)
# Find the month with the highest combined sales.
monthly_total = np.sum(convert_to_nparr,axis=0)
print("monthly total sales : ",monthly_total)
highest_sale_month = np.argmax(monthly_total)
print("month with highest sale : ", highest_sale_month)
# Calculate the average monthly sales for all products combined.
avg_monthly_sales = np.mean(monthly_total)
print("average monthly sales : ", avg_monthly_sales.round(2))
```

Add a new column called Total_Sales.<br>

Summary statistics using .describe()<br>
Find the month where Total_Sales is maximum.<nbr>

```python
months = ["jan","feb","mar","apri","may","jun","july","aug","sep","oct","nov","dec"]
df = pd.DataFrame({"months":months,"product_a" : product_a,"product_b": product_b,"product_c":product_c})

df["total_sales"] = df["product_a"] + df["product_b"] + df["product_c"]
df.describe().head(5)
max_month_idx = df["total_sales"].idxmax()
max_month = df.loc[(max_month_idx)]
max_month
```

```python
plt.plot(df["months"],df["product_a"],label = "product a",marker = "o")
plt.plot(df["months"],df["product_b"],label = "product b",marker = "o")
plt.plot(df["months"],df["product_c"],label = "product c",marker = "o")
plt.title("Product A Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
# total sales of a month
plt.plot(df["months"],df["total_sales"],color="orange",marker = "*")
plt.title("total sales monthly")
plt.xlabel("months")
plt.ylabel("sales")
plt.grid(linestyle=":")
plt.show()
```
