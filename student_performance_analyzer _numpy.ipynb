# Student performance analyzer in numpy

```python
import numpy as np
```

```python



scores = np.array([
    [85, 78, 92, 88, 76],
    [67, 73, 70, 65, 69],
    [95, 89, 94, 98, 91],
    [45, 52, 48, 55, 50],
    [72, 68, 75, 70, 74],
    [88, 90, 85, 92, 87],
    [60, 58, 62, 65, 61],
    [34, 40, 38, 42, 36],
    [79, 81, 77, 80, 78],
    [91, 93, 89, 95, 92]
])
# Each row represents a student, and each column represents a subject.
# Average score per student.
avg_std = np.mean(scores,axis=1)
print("average score per student : ",avg_std)
# Average score per subject.
avg_sub = np.mean(scores ,axis=0)
print("average score of student per subject : ",avg_sub)
# Highest and lowest score in each subject.
high_score = np.max(scores,axis=0)
low_score = np.min(scores,axis=0)
print(f"highest score  in subjects {high_score} and lowest score in subjects {low_score}")
# Students who passed all subjects (≥ 40 in every subject).
passed_Std = np.all(scores >= 40 ,axis=1)
print("passed students : ", passed_Std)
# Students who failed at least one subject.
failed_std = np.any(scores >40 ,axis=1)
print("students who failed atleast one subject : ", failed_std)
# Top 10 students based on average score.
sstd_avg = np.mean(scores ,axis=1)
top_ten_std = np.argsort(sstd_avg)[::-1][:10]
print("top ten students indices : ",top_ten_std)
# Assign grades:
# A: 80+
# B: 70–79
# C: 60–69
# D: 50–59
# F: Below 50
# Display a summary report.
grade = []
for avg in sstd_avg:
    if avg > 80 :
        grade.append("A")
    elif avg > 70 :
        grade.append("B")
    elif avg > 60:
        grade.append("C")
    elif avg> 50 :
        grade.append("D")
    else:
        grade.append("F")
print(grade)
```
