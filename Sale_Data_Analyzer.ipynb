# Sales Data Analyzer (Intermediate)

```python
import numpy as np
```

### row reprsent month and columns represent days

```python
sales = np.random.randint(1000,10000,(12,30))
print(sales)
```

```python
# Total annual revenue
total_revenue = np.sum(sales)
print("total revenue of the annual is : ",total_revenue)
# average annual sales
avg_annual_sales = np.mean(sales)
print("average annual sales : ", avg_annual_sales)
# Maximum sale recorded
max_sale_rec = np.max(sales)
print("maximum sales record : ",max_sale_rec)
# minimum sales record 
min_sales_rec = np.min(sales)
print("minimum sales record  : ", min_sales_rec)
```

## 3. Monthly Analysis

Calculate:<br>
Revenue for each month<br>
Average sales per month<br>
Highest sale in each month<br>
Lowest sale in each month<br>

```python
revenue_per_month = np.sum(sales,axis=1)
print("revenue per months : ", revenue_per_month)
avg_sales_per_month = np.mean(sales,axis=1)
print("average sales per months ", avg_sales_per_month.round(2))
highest_sale_month = np.max(sales,axis=1)
print("highest sales in months : ", highest_sale_month)
lowest_sales_per_month = np.min(sales,axis=1)
print("lowest sale per month is : ", lowest_sales_per_month)
```

## 4. Best and Worst Months

Find:<br>

Month with highest revenue<br>
Month with lowest revenue<br>
Difference between best and worst month<br>

```python
month_highest_revenue = np.argmax(revenue_per_month)
print("month with highest revenue : ",month_highest_revenue)
month_lowest_revenue = np.argmin(revenue_per_month)
print("month with lowest revenue is : ", month_lowest_revenue)
```
