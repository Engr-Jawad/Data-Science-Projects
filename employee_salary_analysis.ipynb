# Project: Company Employee Salary Analysis

You are given salary data for employees in different departments.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python


employee_id = np.arange(1, 21)

department = np.array([
    "IT", "HR", "Finance", "IT", "Marketing",
    "Finance", "HR", "IT", "Marketing", "Finance",
    "IT", "HR", "Marketing", "Finance", "IT",
    "HR", "Marketing", "Finance", "IT", "HR"
])

salary = np.array([
    55000, 48000, 62000, 70000, 51000,
    65000, 47000, 72000, 53000, 68000,
    75000, 50000, 56000, 69000, 80000,
    52000, 59000, 71000, 85000, 54000
])

experience = np.array([
    2, 1, 4, 6, 2,
    5, 1, 7, 3, 5,
    8, 2, 4, 6, 9,
    3, 4, 7, 10, 2
])
```

```python
# highest salary
high_salary = np.max(salary)
print("highest salary =",high_salary)
# lowest salary
lowest_salary = np.min(salary)
print("lowest_salary =",lowest_salary)
# average salary
avg_salary = np.mean(salary)
print("average salary =",avg_salary)

# Employee ID of the highest-paid employee.
high_sal_emp_id = np.argmax(salary)
print("id of  the highest paid employee = ",high_sal_emp_id)
# calculating average years  of experience 
avg_year_exp = np.mean(experience)
print("average years of experience =",avg_year_exp)
```

```python
df = pd.DataFrame({"employees id":employee_id,"department":department,"salary":salary,"experience":experience})
# display first five rows
first_five_rows = df.head(5)
print("displaying first five rows : ", first_five_rows)
print("\n")
# display last three rows
last_three_rows = df.tail(3)
print("last three rows : ",last_three_rows)
print("\n")
# display summary statistics
statistics = df.describe()
print("summary statistics : ",statistics)
print("\n")
#Salary_Category
def Salary_Category(salary):
    if salary >= 70000:
        Salary_Category="High"
    elif (salary >= 55000) and (salary < 70000):
        Salary_Category = "Medium"
    else:
        Salary_Category="Low"
    return Salary_Category
df["Salary_category"] = df["salary"].apply(Salary_Category)
# Find the number of employees in each department.
No_employee_per_dpt = df["department"].value_counts()
print("number of employee per department : ", No_employee_per_dpt)
print("\n")
# find the average salary of each department
avg_salary_per_dpt = df.groupby(department)["salary"].agg(["mean"])
print("average salary of each department : ",avg_salary_per_dpt)

# Find the department with the highest average salary.
dpt_high_avg_salary = df.groupby(department)["salary"].agg(["max","mean"])
dpt_high_avg_salary
```
