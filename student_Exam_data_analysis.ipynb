# Student Exam Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# creating numpy array of math ,science and english

```python
number_of_students = 100
math_marks = np.random.randint(40,101,number_of_students)
science_marks = np.random.randint(40,101,number_of_students)
english_marks = np.random.randint(40,101,number_of_students)
print("math_marks : ",math_marks[:])
print("\n")
print("science_marks : ", science_marks[:])
print("\n")

print("english marks : ", english_marks[:])

```

# creating Dataframe

```python
std_ids=[]
for i in range(1,number_of_students +1):
    std_ids.append(f"S{i:03}")
df = pd.DataFrame({
                      "students_id":std_ids,
                      "math_marks":math_marks,
                      "science_marks":science_marks,
                      "english_marks":english_marks})
df
```

```python
average_marks_maths = df["math_marks"].mean()
print(average_marks_maths)
average_marks_science = df["science_marks"].mean()
print(average_marks_science)
average_marks_english = df["english_marks"].mean()
print(average_marks_english)
print("\n\n")
highest_marks = df[["math_marks","science_marks","english_marks"]].max()
print("highest marks in each subject : \n", highest_marks)
print("\n\n")
lowest_marks = df[["math_marks","science_marks","english_marks"]].min()
print("lowest marks in each subject : \n",lowest_marks)
print("\n\n")
total_marks = df["total_marks"]= df[["math_marks","science_marks","english_marks"]].sum(axis=1)
df["percentage"]= (df["total_marks"]/300 )* 100
def grade(percentage):
    if percentage >= 90 :
        return "A"
    elif percentage >= 80 :
        return "B"
    elif percentage >= 70 :
        return "c"
    elif percentage >= 60:
        return "D"
    else:
        return "F"
df["grade"] = df["percentage"].apply(grade)
df
```

# # top five students on the basis of percentage

```python
top_five = df.sort_values(by="percentage",ascending=False).head(5)
print("top five students : \n",top_five)
```

```python
num_of_grade_a = (df["grade"]== "A").sum()
num_of_grade_f = (df["grade"] == "F").sum()
print("number of students take grade A : ",num_of_grade_a)
print("number of students who failed : ",num_of_grade_f)
```

# # Bar Chart (Average Marks of Each Subject)

```python
subjects = [" math_marks",  "science_marks",  "english_marks "]
average = [df["math_marks"].mean(),df["science_marks"].mean(),df["english_marks"].mean()]
plt.Figure(figsize=(6,4))
bar = plt.bar(subjects,average,color ="green")
plt.bar_label(bar)
plt.xlabel("subjects")
plt.ylabel("average marks")
plt.title("average marks of each subjects")
plt.show()
```

# Histogram (Math Marks Distribution)

```python
plt.Figure(figsize=(8,9))
plt.hist(df["math_marks"],bins=10,edgecolor="black")
plt.title("distribution of math marks ")
plt.xlabel("math_marks")
plt.ylabel("number of students")
plt.show()
```

#  Line Chart (Student ID vs Average Marks)

```python
plt.Figure(figsize=(10,15))
plt.plot(df["students_id"],df["percentage"],marker="o")
plt.title("student average marks")
plt.xlabel("student id")
plt.ylabel("average marks")
plt.xlim(0,13)
plt.show()
```
