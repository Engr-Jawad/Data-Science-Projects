```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/home/jawad/Downloads/netflix_like_10000.csv")
df
```

# cleaning data

```python
df = df.dropna(subset=["type","country","release_year","rating","duration"])
type_count = df["type"].value_counts()
plt.figure(figsize=(8,6))
plt.bar(type_count.index,type_count.values,color=["steelblue","green"],edgecolor="black")
plt.title("Number of Movies vs Tv Show on Netflix")
plt.xlabel("type")
plt.ylabel("count")
plt.tight_layout()
plt.savefig("movies_vs_tv_shows.png",dpi=300,bbox_inches="tight")
plt.show()
```

# content rating distribution

```python
rating_count = df["rating"].value_counts()
plt.figure(figsize=(8,6))
plt.pie(rating_count,labels = rating_count.index,colors=["green","yellow","red","blue"],autopct="%1.1f%%",startangle=90)
plt.title("Percentage of Content Rating")
plt.tight_layout()
plt.savefig("content_rating.png",dpi=200,bbox_inches="tight")
plt.show()




```

```python
movie_df =df[df['type']== 'Movie'].copy()
movie_df["duration_int"] = movie_df["duration"].str.replace("min"," ").astype(int)
plt.figure(figsize=(8,7))
plt.hist(movie_df["duration_int"],bins=20,color="purple",edgecolor="black")
plt.title("Distribution of Movie duration")
plt.xlabel("movie duration")
plt.ylabel("number of movies")
plt.tight_layout()
plt.savefig("movie_distribution_duration.png",dpi=200,bbox_inches ="tight")
plt.show() 

```

```python
relese_count =df["release_year"].value_counts().sort_index()
plt.figure(figsize=(8,6))
plt.scatter(relese_count.index,relese_count.values,color="purple",marker="*")
plt.title("Relese year vs number of shows")
plt.xlabel("relese year")
plt.ylabel("number of shows")
plt.tight_layout()
plt.savefig("release_year_scatter.png",dpi=200,bbox_inches="tight")
plt.show()
```

```python
# top ten country 
country_count = df["country"].value_counts().head(10)
plt.figure(figsize=(8,7))
plt.bar(country_count.index,country_count.values,color="teal")
plt.title("top ten countries by number of shows")
plt.xlabel("number of shows")
plt.ylabel("country")
plt.xticks(rotation = 45)
plt.tight_layout()
plt.savefig("top_ten_country_bar.png")
plt.show()
```
